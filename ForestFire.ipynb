## Graphs and Observations

```python
sns.relplot(kind = 'scatter',x ='month',y = 'day',data=FF,height=6,aspect=2,size = 'FFMC',)
```

It shows how the FFMC is increasing and decreasing.At some point like in month of "jan" day "sun" FFMC is depleting

```python
sns.relplot(kind = 'scatter',x ='month',y = 'day',data=FF,height=6,aspect=2,size = 'DMC')
```

As you can see in the Above Graph.The Bigger dots represents the increament and the small dots represent the decreament in DMC

```python
sns.relplot(kind = 'scatter',x ='month',y = 'day',data=FF,size = 'ISI',height=6,aspect=2)
```

It also shows the increase and decrease rate in ISI as the month and day is passing.On month of "jun" day "sun" there is an increament in ISI

#### As you can see here in this graph when you check the month of ,Sep,Aug,Jul and  DC is in increasing rate

```python
sns.relplot(kind = 'scatter',x ='month',y = 'day',data=FF,size= 'DC',height=6,aspect=2)
```

But at some points DC is also decreasing like you can check in the above graph after feb 

### Almost every month shows similar things as "FFMC" or "DMC" or "DC" or "ISI" is Increasing or Decreasing per Day

```python
sns.relplot(x='temp',y='area',data=FF,kind='line');
```

This above graph show increasing in temperature indirectly results in area depletion.

```python
sns.relplot(x=sorted(FF.month),y='area',data=FF,kind='line');
```

Same observation of depletion of area over breif period of months.

```python
sns.lineplot(x='rain',y='RH',data=FF);
```

This graph shows relation of humidity with rain, as humidity increases rain start and with continous rain humidity decreases. 


```python
sns.lineplot(x='rain',y='RH',data=FF,hue='month',sort=True)

```

This graph shows relation between Humidity and Rainfall


#### Observation 

Almost every month shows smillar relation, as humidity increases causes rainfall but in month of August rainfall increases within range of 70-60 RH level.  


```python
sns.lineplot(x='wind',y='temp',data=FF)
```

This graph shows as speed of wind increases temperature tends to drop at some point.

```python
sns.barplot(x='month',y='temp',data=FF)
```

This graph simply shows variation of temperature over period of months.
